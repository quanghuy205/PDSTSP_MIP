## Distribuciones continuas  <a class="anchor" id="continuas"></a>
### Exponencial <a class="anchor" id="exp"></a>
$
f_X(x;\lambda) = \begin{cases}
                     \lambda e^{-\lambda x} & x \geq 0 \\
                     0 & x<0
                 \end{cases}
$

```python
# par√°metro lambda
lam = 0.2

# definir variable aleatoria
x = st.expon(scale=1/lam, loc=0)

# valor del eje x para graficar
val = np.linspace(0, 40, 1000)

# figura que tendr√° los dos gr√°ficos
fig = plt.figure(figsize=(15, 5))

# gr√°fico de la distribuci√≥n (pdf)
ax = fig.add_subplot(121)
ax.set_title('pdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$f_X(x)$')
ax.plot(val, x.pdf(val))             # usar pdf para graficar

# gr√°fico de la distribuci√≥n (cdf)
ax = fig.add_subplot(122)
ax.set_title('cdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$F_X(x)$')
ax.plot(val, x.cdf(val))             # usar cdf para graficar
```

### Normal <a class="anchor" id="normal"></a>
$
f_X(x;\mu,\sigma^2) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{\left(x-\mu\right)^2}{2\sigma^2}}
$

```python
# par√°metro mu y sigma
mu = 10
sigma = 4

# definir variable aleatoria
x = st.norm(loc=mu, scale=sigma)

# valor del eje x para graficar
val = np.linspace(-5, 25, 1000)

# figura que tendr√° los dos gr√°ficos
fig = plt.figure(figsize=(15, 5))

# gr√°fico de la distribuci√≥n (pdf)
ax = fig.add_subplot(121)
ax.set_title('pdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$f_X(x)$')
ax.plot(val, x.pdf(val))             # usar pdf para graficar

# gr√°fico de la distribuci√≥n (cdf)
ax = fig.add_subplot(122)
ax.set_title('cdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$F_X(x)$')
ax.plot(val, x.cdf(val))             # usar cdf para graficar
```

### Uniforme / Rectangular <a class="anchor" id="uniforme"></a>
$
f_X(x;a,b) = \begin{cases}
                     \frac{1}{b-a} & a\leq x \leq b \\
                      0 & \text{d.o.f.}
                 \end{cases}
$

```python
# par√°metros a y b
a = 8
b = 22

# definir variable aleatoria
x = st.uniform(loc=a, scale=b-a)

# valor del eje x para graficar
val = np.linspace(0, 30, 1000)

# figura que tendr√° los dos gr√°ficos
fig = plt.figure(figsize=(15, 5))

# gr√°fico de la distribuci√≥n (pdf)
ax = fig.add_subplot(121)
ax.set_title('pdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$f_X(x)$')
ax.plot(val, x.pdf(val))             # usar pdf para graficar

# gr√°fico de la distribuci√≥n (cdf)
ax = fig.add_subplot(122)
ax.set_title('cdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$F_X(x)$')
ax.plot(val, x.cdf(val))             # usar cdf para graficar
```

### Triangular  <a class="anchor" id="tri"></a>
$
f_X(x;a,b,c) = \begin{cases}
                     \frac{2(x-a)}{(b-a)(c-a)} & a\leq x \leq c \\
                     \frac{2(b-x)}{(b-a)(b-c)} & c\leq x \leq b \\
                      0 & \text{d.o.f.}
                 \end{cases}
$

```python
# par√°metros a, b y c
a = 5
b = 25
c = 10

# definir variable aleatoria
x = st.triang((c-a)/(b-a), loc=a, scale=b-a)

# valor del eje x para graficar
val = np.linspace(0, 30, 1000)

# figura que tendr√° los dos gr√°ficos
fig = plt.figure(figsize=(15, 5))

# gr√°fico de la distribuci√≥n (pdf)
ax = fig.add_subplot(121)
ax.set_title('pdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$f_X(x)$')
ax.plot(val, x.pdf(val))             # usar pdf para graficar

# gr√°fico de la distribuci√≥n (cdf)
ax = fig.add_subplot(122)
ax.set_title('cdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$F_X(x)$')
ax.plot(val, x.cdf(val))             # usar cdf para graficar
```

### Trapezoidal  <a class="anchor" id="trap"></a>
$
f_X(x;a,b,c) = \begin{cases}
                     \frac{2}{b-a+m_2-m_1}\frac{x-a}{m_1-a} & a\leq x \leq m_1 \\
                     \frac{2}{b-a+m_2-m_1} & m_1\leq x \leq m_2 \\
                     \frac{2}{b-a+m_2-m_1}\frac{b-x}{b-m_2} & m_2\leq x \leq b \\
                      0 & \text{d.o.f.}
                 \end{cases}
$

```python
# par√°metros a, b, m1 y m2
a = 2
b = 25
m1 = 6
m2 = 15

# definir variable aleatoria
x = st.trapz((m1-a)/(b-a), (m2-a)/(b-a), loc=a, scale=b-a)

# valor del eje x para graficar
val = np.linspace(0, 30, 1000)

# figura que tendr√° los dos gr√°ficos
fig = plt.figure(figsize=(15, 5))

# gr√°fico de la distribuci√≥n (pdf)
ax = fig.add_subplot(121)
ax.set_title('pdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$f_X(x)$')
ax.plot(val, x.pdf(val))             # usar pdf para graficar

# gr√°fico de la distribuci√≥n (cdf)
ax = fig.add_subplot(122)
ax.set_title('cdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$F_X(x)$')
ax.plot(val, x.cdf(val))             # usar cdf para graficar
```

### Gamma   <a class="anchor" id="gamma"></a>
$
f_X(x;\alpha,\beta) =\begin{cases}
                     \frac{1}{\beta^\alpha\Gamma(\alpha)} & x \geq 0 \\
                      0 & \text{d.o.f.}
                 \end{cases}
$

```python
# par√°metros alfa y beta
a = 4
b = 2.5

# definir variable aleatoria
x = st.gamma(a, loc=0, scale=b)

# valor del eje x para graficar
val = np.linspace(0, 40, 1000)

# figura que tendr√° los dos gr√°ficos
fig = plt.figure(figsize=(15, 5))

# gr√°fico de la distribuci√≥n (pdf)
ax = fig.add_subplot(121)
ax.set_title('pdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$f_X(x)$')
ax.plot(val, x.pdf(val))             # usar pdf para graficar

# gr√°fico de la distribuci√≥n (cdf)
ax = fig.add_subplot(122)
ax.set_title('cdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$F_X(x)$')
ax.plot(val, x.cdf(val))             # usar cdf para graficar
```

### Beta <a class="anchor" id="beta"></a>
$
f_X(x;\alpha,\beta,a,b) =
\begin{cases}
    \frac{1}{\text{B}(\alpha,\beta)(b-a)}
        \left(\frac{x-a}{b-a}\right)^{\alpha-1}
        \left(\frac{b-x}{b-a}\right)^{\beta-1}
        & a \leq x \leq b \\
    0 & \text{d.o.f.}
\end{cases}
$

```python
# par√°metros a, b, alfa, beta
a = -1
b = 5
alfa = 0.8
beta = 0.8

# verificar par√°metros
assert b > a and alfa > 0 and beta > 0, 'Par√°metros no v√°lidos.'

# definir variable aleatoria
x = st.beta(alfa, beta, loc=a, scale=b-a)

# valor del eje x para graficar
val = np.linspace(a, b, 1000)

# figura que tendr√° los dos gr√°ficos
fig = plt.figure(figsize=(15, 5))

# gr√°fico de la distribuci√≥n (pdf)
ax = fig.add_subplot(121)
ax.set_title('pdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$f_X(x)$')
ax.plot(val, x.pdf(val))             # usar pdf para graficar

# gr√°fico de la distribuci√≥n (cdf)
ax = fig.add_subplot(122)
ax.set_title('cdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$F_X(x)$')
ax.plot(val, x.cdf(val))             # usar cdf para graficar
```

## Distribuciones discretas  <a class="anchor" id="discretas"></a>
### Binomial <a class="anchor" id="binomial"></a>

```python
# par√°metros n y p
n = 10
p = 0.4

# verificar par√°metros
assert n > 0 and isinstance(n, int) and 0 < p < 1, 'Par√°metros no v√°lidos.'

# definir variable aleatoria
x = st.binom(n, p)

# figura que tendr√° los dos gr√°ficos
fig = plt.figure(figsize=(15, 5))

# gr√°fico de la distribuci√≥n (pmf)
ax = fig.add_subplot(121)
ax.set_title('pmf')
ax.set_xlabel('$x$')
ax.set_ylabel('$p_X(x)$')
val = np.linspace(0, n, n+1)                          # valor del eje x para graficar
plt.plot(val, x.pmf(val), ls='None', marker='o')      # usar pmf para graficar

# gr√°fico de la distribuci√≥n (cdf)
ax = fig.add_subplot(122)
ax.set_title('cdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$F_X(x)$')
val = np.linspace(-1, n + 1, 1000)                    # valor del eje x para graficar
ax.plot(val, x.cdf(val))                              # usar cdf para graficar
```

```python
# par√°metros lambda
lam = 4

# definir variable aleatoria
x = st.poisson(lam)

# figura que tendr√° los dos gr√°ficos
fig = plt.figure(figsize=(15, 5))

# gr√°fico de la distribuci√≥n (pmf)
ax = fig.add_subplot(121)
ax.set_title('pmf')
ax.set_xlabel('$x$')
ax.set_ylabel('$p_X(x)$')
val = np.linspace(0, 14, 15)                          # valor del eje x para graficar
plt.plot(val, x.pmf(val), ls='None', marker='o')      # usar pmf para graficar

# gr√°fico de la distribuci√≥n (cdf)
ax = fig.add_subplot(122)
ax.set_title('cdf')
ax.set_xlabel('$x$')
ax.set_ylabel('$F_X(x)$')
val = np.linspace(-1, 15, 1000)                       # valor del eje x para graficar
ax.plot(val, x.cdf(val))                              # usar cdf para graficar
```
