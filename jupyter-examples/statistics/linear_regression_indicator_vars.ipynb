```python
import pandas as pd
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_excel('table 12-3.xlsx')
df['nv'] = df['n/v']
df.head()
```

```python
modelo = smf.ols('mpg ~ cid + rhp + etw + cmp + axle + nv + C(trns)', data=df)
resultado = modelo.fit()
resultado.summary()
```

```python
predic = resultado.get_prediction(exog={'cid': 215,
                                        'rhp': 253,
                                        'etw': 4500,
                                        'cmp': 9.9,
                                        'axle': 3.07,
                                        'nv': 30.9,
                                        'trns': 'M6'})
```
